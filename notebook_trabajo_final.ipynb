```python
import pandas as pd
```

```python
from faker import Faker
fake = Faker()
```

```python
lista = []
```

```python
for x in range(0,10001):
    lista.append({'name': fake.name(),'last_name': fake.last_name(),'email': fake.email(),'address': fake.address()})
```

```python
lista
```

```python
data = pd.DataFrame(lista)
```

```python
def add_edad(d):
    edad = fake.random_int(min=30, max=40)
    return edad
```

```python
data['age'] = data['age'].apply(add_edad)
```

```python
def add_provincia(d):
    provincias = ["Loja", "Pichincha", "Guayas", "Azuay", "Manabí"]
    return fake.random_element(provincias)
```

```python
data['province'] = data['province'].apply(add_provincia)
```

```python
data.to_csv("misdata.csv", index=False)
```
